```python
import matplotlib.pyplot as plt
#download so we can create graph 
```

```python
def N_exp(N_12, p, time):
    # exponential growth model
    # save N_values into a list
    N_values=[N_12]
    # save initial population
    current_N=N_12
    # set curent n value
    N0 = 12
    b = 1.0
    d = 0.4
    p = b - d  # 0.6
    time = 10  # years to project
    for _ in range(time):
        # set up abaility to iterate through time
        population_projection = N_exp(N0, p, time)
    for year, pop in enumerate(population_projection):
        print(f"Year {year}: {pop} plants")
        current_N = current_N + p * current_N
        # exponential growth equation
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        # rounds to three decimal places
        N_values.append(current_N)
        # lists n values
        # starts at current n
    return N_values
     # puts n values into a list
```

```python
def N_log(N_0, p, time, stop_at=None):
    N_values = [N_0]
    current_N = N_0
    for _ in range(time):
        current_N = round(current_N * (1 + p), 3)
        N_values.append(current_N)
        if stop_at is not None and current_N >= stop_at:
            break
    return N_values

# Parameters
N_0 = 12          # initial population
b = 1.0          # birth rate
d = 0.4          # death rate
p = b - d        # net growth rate
time = 100       # max years to project
stop_population = 400  # stop condition

population_projection = N_log(N_0, p, time, stop_at=stop_population)

for year, pop in enumerate(population_projection):
    print(f"Year {year}: {pop} plants")
#run the funtion with 35 days and p = 0.4 with an initial population of 0.01
```

```python
results = N_log(12, 0.4, 400, stop_at=400)

print(results)
# prints out population each year
```

```python
# plot the results with days on the x axis and population size on the y axis

year = list(range(8))  # 0 to 8
#sets a limit of 35 iterations = 0-8 years
plt.plot(year, results, marker ='o')
# plots the results with dots at each point
plt.xlabel('year')
# x axis label
plt.ylabel('Population')
# y axis label
plt.title('Population Growth Over Time')
# title of the graph
plt.grid()
# adds a grid to the graph
plt.show()
#puts it all together and creates graph
```
