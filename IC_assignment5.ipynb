```python
# Normal distribution example: Leaf length (cm) in a tree species
# Author: [Your Name]
# Description: Simulates normally distributed ecological data and plots it.

import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Generate simulated data ---
np.random.seed(42)  # for reproducibility

mean_leaf_length = 10.0   # mean (μ) in cm
std_dev = 2.0             # standard deviation (σ)
sample_size = 100         # number of leaves

# Generate random data following a normal distribution
leaf_lengths = np.random.normal(mean_leaf_length, std_dev, sample_size)

# --- Step 2: Print summary statistics ---
print(f"Mean leaf length: {np.mean(leaf_lengths):.2f} cm")
print(f"Standard deviation: {np.std(leaf_lengths):.2f} cm")
print(f"Sample size: {len(leaf_lengths)}")

# --- Step 3: Visualize the distribution ---
plt.figure(figsize=(8, 5))
plt.hist(leaf_lengths, bins=10, color='forestgreen', edgecolor='black', alpha=0.7, density=True)
plt.title("Normal Distribution of Leaf Lengths in Red Oak (Quercus rubra)")
plt.xlabel("Leaf length (cm)")
plt.ylabel("Probability Density")

# Overlay a smooth normal curve for reference
x = np.linspace(leaf_lengths.min(), leaf_lengths.max(), 100)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_leaf_length) / std_dev) ** 2)
plt.plot(x, pdf, 'r--', label='Theoretical Normal Curve')
plt.legend()

plt.show()
```
