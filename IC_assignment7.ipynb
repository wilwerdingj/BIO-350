```python
import matplotlib.pyplot as plt
import numpy as np
#download so we can create graph
```

```python
A = np.array([
    [0.8,  0.25, 0.1],   # Beech
    [0.1,  0.15, 0.25],   # Maple
    [0.1,  0.6, 0.65]    # Other
])

# Initial state: [Beech, Maple, Other]
n0 = np.array([1, 20, 79])

# Simulate over 50 time steps
years = 100
pop = [n0]

for _ in range(years):
    n_next = A @ pop[-1]
    pop.append(n_next)

pop = np.array(pop)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(pop[:, 0], label="Beech")
plt.plot(pop[:, 1], label="Maple")
plt.plot(pop[:, 2], label="Other")
plt.title("Forest Succession Dynamics (Beech–Maple–Other)")
plt.xlabel("Time (years)")
plt.ylabel("Relative Abundance")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
final_pop =  pop[-1] .round() .astype(int)
print("Final population distribution after", years, "years:")
print("Beech:", final_pop[0])
print("Maple:", final_pop[1])
print("Other:", final_pop[2])
```

```python

n = np.array([0, 0, 0, 1], dtype=float)

A = np.array([
    [1, 0.029, 0.13, 0.0039],   
    [0, 0.95, 0.0026, 0.13],  
    [0, 0.00060, 0.85, 0.026],
    [0, 0.019, 0.017, 0.84]    
])

timesteps = 50
history = [n.copy()]

for t in range(timesteps):
    n_next = A @ n
    history.append(n_next)
    n = n_next


history = np.array(history)

years = np.arange(timesteps + 1)
plt.figure(figsize=(8, 5))
plt.plot(years, history[:, 0], label='Extinct')
plt.title('Butterfly Population Dynamics')
plt.xlabel('Time Step (Years)')
plt.ylabel('Probability')
plt.grid(True)
plt.ylim(0, 1)
plt.legend()
```

```python
#calculate the probability of extinction 
final_extinction_probability = history[-1, 0]  
print("Probability of extinction after", timesteps, "years:", final_extinction_probability)
```
