```python
import random
from typing import Tuple

def total_nests(n1: int, n2: int, n3: int) -> int:
    return n1 + n2 + n3

def avg_eggs_per_nest(n1: int, n2: int, n3: int) -> float:
    N = total_nests(n1, n2, n3)
    if N == 0:
        return 0.0
    return (1 * n1 + 2 * n2 + 3 * n3) / N

def prob_first_is(n1: int, n2: int, n3: int, target: int) -> float:
    """Probability the first nest checked has 'target' eggs."""
    N = total_nests(n1, n2, n3)
    if N == 0:
        return 0.0
    counts = {1: n1, 2: n2, 3: n3}
    return counts.get(target, 0) / N

def prob_k_in_a_row(n1: int, n2: int, n3: int, target: int, k: int, replace: bool = False) -> float:
    """
    Probability of drawing k nests in a row that each have 'target' eggs.
    By default assumes sampling without replacement (replace=False).
    If replace=True, draws are independent and probability is (p)^k.
    """
    counts = {1: n1, 2: n2, 3: n3}
    N = total_nests(n1, n2, n3)
    target_count = counts.get(target, 0)
    if k <= 0:
        return 1.0
    if N == 0:
        return 0.0
    if replace:
        p = target_count / N
        return p ** k
    # without replacement
    if target_count < k:
        return 0.0
    prob = 1.0
    for i in range(k):
        prob *= (target_count - i) / (N - i)
    return prob

def simulate_probabilities(n1: int, n2: int, n3: int, target: int, k: int,
                           trials: int = 20000, replace: bool = False, seed: int = None) -> Tuple[float, float]:
    """
    Monte Carlo simulation to estimate:
      - probability first nest has 'target' eggs
      - probability first k draws are all 'target' eggs
    Returns (p_first_estimate, p_k_in_a_row_estimate)
    """
    if seed is not None:
        random.seed(seed)
    pool = [1] * n1 + [2] * n2 + [3] * n3
    if len(pool) == 0:
        return 0.0, 0.0
    first_count = 0
    k_count = 0
    for _ in range(trials):
        if replace:
            # draw first separately to always have a 'first'
            first = random.choice(pool)
            if k <= 0:
                draws = []
            else:
                draws = [first] + [random.choice(pool) for _ in range(k - 1)]
        else:
            # create a random permutation of the pool and take the first k (pad with None if k > len(pool))
            perm = random.sample(pool, len(pool))
            first = perm[0]
            if k <= len(pool):
                draws = perm[:k]
            else:
                draws = perm + [None] * (k - len(pool))
        if first == target:
            first_count += 1
        # follow the convention that k <= 0 means the k-in-a-row event is trivially true
        if k <= 0:
            k_count += 1
        else:
            if len(draws) >= k and all(d == target for d in draws[:k]):
                k_count += 1
    return first_count / trials, k_count / trials

# Example usage:
# Adjust these input parameters as needed.
n1, n2, n3 = 50, 30, 20   # number of nests with 1, 2, 3 eggs
target = 2                # eggs to check for the first-nest probability
k = 3                     # how many nests in a row to check for 'target' eggs

# Exact computations (without replacement by default)
p_first = prob_first_is(n1, n2, n3, target)
p_k_row = prob_k_in_a_row(n1, n2, n3, target=1, k=3, replace=False)  # example: 3 nests in a row with 1 egg
average = avg_eggs_per_nest(n1, n2, n3)

print(f"Counts: 1-egg={n1}, 2-egg={n2}, 3-egg={n3} (total={total_nests(n1,n2,n3)})")
print(f"Probability first nest has {target} eggs: {p_first:.6f}")
print(f"Probability of 3 nests in a row with 1 egg (no replacement): {p_k_row:.6f}")
print(f"Average eggs per nest: {average:.4f}")

# Quick simulation to validate formulas (optional)
p_first_sim, p_k_sim = simulate_probabilities(n1, n2, n3, target=1, k=3, trials=20000, replace=False, seed=42)
print(f"Simulated P(first is 1): {p_first_sim:.4f}, Simulated P(3 in a row of 1): {p_k_sim:.4f}")
```
