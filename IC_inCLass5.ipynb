Creating SIR model over a 30 day span

```python
import matplotlib.pyplot as plt
import numpy as np
#download so we can create graph
```

```python
def sir_model(S0, I0, R0, b, d_rate, day):
    #set constants and initial values
    S0 = 999
    I0 = 1
    R0 = 0
    b = 0.001
    d_rate = 0.5
    S = [S0]
    I = [I0]
    R = [R0]
    for i in range(day):
        S_new = S[-1] - (b * S[-1] * I[-1])
        I_new = I[-1] + (b * S[-1] * I[-1]) - (d_rate * I[-1])
        R_new = R[-1] + (d_rate * I[-1])
        S.append(S_new)
        I.append(I_new)
        R.append(R_new)
    return S, I, R
        
```

```python
resultS, resultI, resultR = sir_model(999, 1, 0, 0.5, 0.001, 30)
print(resultS, resultI, resultR)
```

```python
# Plot Npop and Mpop over days, each on its own line
days = list(range(31))  # Days from 0 to 35
plt.plot(days, resultS, color='blue', marker='o')
plt.plot(days, resultI, color='orange', marker='o')
plt.plot(days, resultR, color='red', marker='o')
plt.xlabel('Days')
plt.ylabel('Fox , Suseptable, Infected, Recovered')
plt.title('Rabies infection in Fox Population')
plt.grid()
plt.legend()
plt.show()
```
