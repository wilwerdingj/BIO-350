```python
import matplotlib.pyplot as plt
#download so we can create graph 
```

```python
def N_exp(N_0, p, time):
    # exponential growth model
    # save N_values into a list
    N_values=[N_0]
    # save initial population
    current_N=N_0
    # set curent n value
    for _ in range(time):
        # set up abaility to iterate through time
        current_N = current_N + p * current_N
        # exponential growth equation
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        # rounds to three decimal places
        N_values.append(current_N)
        # lists n values
        # starts at current n
    return N_values
     # puts n values into a list
```

```python
#Call N_iterative with N_0 = 0, p=2, number_iterations = 10
results = N_exp(0.01, 0.4 , 35)
# and print the results
print(results)
```

```python
days = list(range(36))  # 0 to 35
#sets a limit of 35 iterations = 0-35 days
plt.plot(days, results,marker='o')
# plots the results with dots at each point
plt.xlabel('Days')
# x axis label
plt.ylabel('Duckweed Population')
# y axis label
plt.title('Duckweed Population Growth Over Time')
# title of the graph
plt.grid()
# adds a grid to the graph
plt.show()
#puts it all together and creates graph
```
