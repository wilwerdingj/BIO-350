```python
#Import "Population_data.csv" into a pandas dataframe 
import pandas as pd
pop_data = pd.read_csv("Population_Data (2).csv")
pop_data.head()  # Display the first few rows of the dataframe
```

```python
# average the population from day 17 to day 60 
avg_pop = pop_data[(pop_data["Days"] >= 17) & (pop_data["Days"] <= 60)]["Population"].mean()
print(f"Average population from day 17 to day 60: {avg_pop}")
```

```python
import matplotlib.pyplot as plt

plt.plot(pop_data["Days"], pop_data["Population"], marker='o', color='black', linestyle='-')
plt.axhline(y=276, color='green', linestyle='--', linewidth=2, label='Carrying Capacity')
plt.xlabel("Days")
plt.ylabel("Population")
plt.title("Population Growth Over Time")
plt.xlim(left=0)
plt.ylim(bottom=0)

# Add y=276 to the y-axis ticks and sort them
yticks = plt.yticks()[0].tolist()
if 276 not in yticks:
    yticks.append(276)
    yticks = sorted(yticks)
plt.yticks(yticks)

plt.legend()
plt.show()
```
