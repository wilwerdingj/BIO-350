```python
import matplotlib.pyplot as plt
#download so we can create graph 
```

```python
def NandM_population(N_0, M_0, time, K_N, K_M, d_N, d_M, S):
    #growth model
    #save N_values and M_values into a list
    N_values = [N_0]
    M_values = [M_0]
    #saves inital values to list
    current_N = N_0
    current_M = M_0
    #set current population size
    # N_0 = 5
    # M_0 = 12
    # S = 1
    # d_N = 0.4
    # d_M = 0.6
    # K_N = 400
    # K_M = 1600
    for _ in range(time):
        current_N = current_N + S * current_N * (1 - (current_N / K_N) - (current_M / K_M)) - d_N * current_N
        current_M = current_M + S * current_M * (1 - (current_M / K_M) - (current_N / K_N)) - d_M * current_M
        current_M = round(current_M, 3)
        M_values.append(current_M)
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values, M_values
#returns the list of population sizes over time
```

```python
#run the funtion with 35 days and p = 0.4 with an initial population of 0.01
#def NandM_population(current_N, current_M, time, K_N, K_M, d_N, d_M, S):
Npop,Mpop = NandM_population(12,5, 35, 400, 1600, 0.4, 0.6, 1)
# initial values
print(Npop, Mpop)
# gives the results in a readable format
```

```python
# Plot Npop and Mpop over days, each on its own line
days = list(range(36))  # Days from 0 to 35
plt.plot(days, Npop, label='Species N Population', color='blue', marker='o')
plt.plot(days, Mpop, label='Species M Population', color='orange', marker='o')
plt.xlabel('Days')
plt.ylabel('Population')
plt.title('Lotka-Volterra Competition: Population Over Time')
plt.grid()
plt.legend()
plt.show()
```
